```python
import plotly.graph_objects as go
```

```python
df_stocks = px.data.stocks()
fig = px.line(df_stocks,x="date",y=["GOOG","AAPL"],labels={'x':'date','y':'price'},title="prices GOOGLE VS APPLE")
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_stocks.date,y=df_stocks.GOOG,mode='lines',name='google'))
fig.add_trace(go.Scatter(x=df_stocks.date,y=df_stocks.AAPL,mode='lines+markers',name='apple'))
```

```python
df_us = px.data.gapminder().query("country == 'United States'")
px.bar(df_us,x='year',y='pop')
df_tips = px.data.tips()
px.bar(df_tips,x='day',y='tip',color="sex",title='tips by sex on each day',
      labels={'tip':'tip amount','day':'days of week'})
```

```python
px.bar(df_tips,x='sex',y='total_bill',color="smoker",barmode="group")
```

```python
df_europe = px.data.gapminder().query("continent == 'Europe' and year == 2007 and pop > 2.e6")
fig = px.bar(df_europe,y='pop',x='country',text="pop",color='country')
fig.show()
```

```python
df_europe = px.data.gapminder().query("continent == 'Europe' and year == 2007 and pop > 2.e6")
fig = px.bar(df_europe,y='pop',x='country',text="pop",color='country')
fig.update_traces(texttemplate='%{text:.2s}',textposition='outside')
fig.show()
```
